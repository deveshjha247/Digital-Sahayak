```python
# Setup
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import json
import pandas as pd
from ai.config import config, get_hyperparams

print("‚úÖ Setup complete!")
```

## 1. Load Training Data

```python
# Load intent samples
data_path = Path.cwd().parent / "ai" / "data" / "raw" / "intent_samples.json"

with open(data_path, 'r', encoding='utf-8') as f:
    intent_data = json.load(f)

df = pd.DataFrame(intent_data)
print(f"üìä Loaded {len(df)} samples")
print(f"\nüìã Intent Distribution:")
print(df['intent'].value_counts())
```

## 2. Load Hyperparameters from Config

```python
# Load hyperparameters from config file (NOT hardcoded!)
hp = get_hyperparams("intent_classifier")

EPOCHS = hp['epochs']
BATCH_SIZE = hp['batch_size']
LEARNING_RATE = hp['learning_rate']
WEIGHT_DECAY = hp['weight_decay']
DROPOUT = hp['model_config']['dropout']

print("üîß Hyperparameters loaded from config:")
print(f"  Epochs: {EPOCHS}")
print(f"  Batch Size: {BATCH_SIZE}")
print(f"  Learning Rate: {LEARNING_RATE}")
print(f"  Weight Decay: {WEIGHT_DECAY}")
print(f"  Dropout: {DROPOUT}")
```

## 3. View Sample Data

```python
# View sample data
print("üìù Sample Training Data:\n")
for intent in df['intent'].unique()[:5]:
    sample = df[df['intent'] == intent].iloc[0]
    print(f"Intent: {intent}")
    print(f"  Text: {sample['text']}")
    print(f"  Language: {sample['language']}")
    print()
```
